```python
import torch

def quat_rotate_inverse(q, v):
    shape = q.shape
    q_w = q[:, -1]
    q_vec = q[:, :3]
    
    a = v * (2.0 * q_w ** 2 - 1.0).unsqueeze(-1)
    b = torch.cross(q_vec, v, dim=-1) * q_w.unsqueeze(-1) * 2.0
    c = q_vec * torch.bmm(q_vec.view(shape[0], 1, 3), v.view(shape[0], 3, 1)).squeeze(-1) * 2.0
    
    return a - b + c

# Example usage
q = torch.tensor([
    [0.0, 0.0, 0.0, 1.0],     # Identity quaternion (no rotation)
    [0.6, 0.0, 0.6, 0.0] # 90-degree rotation around the y-axis
])
v = torch.tensor([
    [1.0, 0.0, 0.0], # Vector along the x-axis
    [0.0, 1.0, 0.0]  # Vector along the y-axis
])

rotated_v = quat_rotate_inverse(q, v)
print(rotated_v)
```
